```python
## Import the neccessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
# Now lets read the facebook dataset
data = pd.read_csv('fb_price.csv')
```

```python
## Lets now print the first five rows of the dataset
data.head()
```

```python
## Print the last five rows of the dataset
data.tail()
```

```python
data.columns
```

```python
## Lets select the date and Adjusted Prices only
data = data[['Date','Adj Close']]
```

```python
## Print the first five rows
data.head()
```

```python
### Lets rename the columns
data.rename(columns={"Date":'date','Adj Close':'price_t'},inplace=True)
```

```python
data.head()
```

```python
# Lets calcuate the returns 
data['returns'] = (data['price_t']/data['price_t'].shift(1))-1
```

```python
data.head()
```

```python
### Alternative approach
data['return1'] = data['price_t'].pct_change(1)
```

```python
# Code to set the index to dates
data.set_index('date',inplace=True)
```

```python
data.head()
```

```python
## We need to plot the prices over certain dates
data['price_t'].plot(figsize=(10,8),xlabel='Dates',ylabel='Stock Prices($)',title='The Facebook Stock Prices');
```

```python
## We need to plot the returns over certain dates
data['returns'].plot(figsize=(10,8),xlabel='Dates',ylabel='Returns',title='The Facebook Returns');
```
